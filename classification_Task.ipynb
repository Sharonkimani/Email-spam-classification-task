```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split , cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

```

```python
data= pd.read_csv('/content/emails.csv')
data.head()
```

```python
data.info(memory_usage='deep')
data.drop('Email No.', axis=1, inplace=True)
print(data.info(memory_usage='deep'))
```

```python
print(f'The dataset has {data.duplicated().sum()} duplicates')
print(f'The dataset has {data.isna().sum().sum()} missing values')
```

```python
data[data.duplicated(keep=False)]
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
data.shape
```

# Modelling


```python
y =data[['Prediction']]
X= data.drop('Prediction', axis=1)
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state=24)
```

```python
# standardizing
scaler= StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
y_train.shape
```

```python
# baseline model
model1= LogisticRegression()
model1.fit(X_train_scaled, y_train)
model1_pred= model1.predict(X_test_scaled)
# Evaluate Logistic Regression
print("Logistic Regression Accuracy:", accuracy_score(y_test, model1_pred))
print("Logistic Regression Classification Report:\n", classification_report(y_test, model1_pred))
print("Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test, model1_pred))
```

```python
# decision tree classifier
model2= DecisionTreeClassifier()
model2.fit(X_train_scaled, y_train)
model2_pred= model2.predict(X_test_scaled)
# Evaluate the decision tree classifier
print("Decision Tree Classifier Accuracy:", accuracy_score(y_test, model2_pred))
print("Decision Tree Classifier Classification Report:\n", classification_report(y_test, model2_pred))
print("Decision Tree Classifier Confusion Matrix:\n", confusion_matrix(y_test, model2_pred))
```

```python
#Random forest classifer
model3= RandomForestClassifier()
model3.fit(X_train_scaled, y_train)
model3_pred= model3.predict(X_test_scaled)
# Evaluate the random forest classifier
print("Random forest Classifier Accuracy:", accuracy_score(y_test, model3_pred))
print("Random forest Classifier Classification Report:\n", classification_report(y_test, model3_pred))
print("Random forest Classifier Confusion Matrix:\n", confusion_matrix(y_test, model3_pred))
```

The best performing model is the random forest classifier. thus that is our chosen model.

```python

# Define the parameter grid to search
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create the Random Forest model
rf_model = RandomForestClassifier()

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid,
                           cv=3, scoring='accuracy')

# Fit the model to the data
grid_search.fit(X_train_scaled, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Use the best model for predictions
best_rf_model = grid_search.best_estimator_
y_pred_rf_tuned = best_rf_model.predict(X_test_scaled)

# Evaluate the tuned Random Forest Classifier
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
print("Tuned Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf_tuned))
print("Tuned Random Forest Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_tuned))
```

After hyperparameter tuning, the accuracy reduced , using the baseline random forest classifier is the best way to go.
